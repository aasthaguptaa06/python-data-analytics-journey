```
# Week 2 - For Loops Deep Dive
# Author: Aastha Gupta

# Looping with range - different forms
print("range(5):")
for i in range(5):
    print(i, end=" ")

print("\n\nrange(2, 10):")
for i in range(2, 10):
    print(i, end=" ")

print("\n\nrange(0, 20, 2) - step of 2:")
for i in range(0, 20, 2):
    print(i, end=" ")

print("\n\nrange(10, 0, -1) - counting backwards:")
for i in range(10, 0, -1):
    print(i, end=" ")

# Looping over a list with enumerate
print("\n\nLooping with enumerate:")
fruits = ["mango", "banana", "apple", "guava"]
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")

# Looping over two lists together with zip
names = ["Aastha", "Riya", "Karan"]
marks = [85, 78, 92]
print("\nLooping with zip:")
for name, mark in zip(names, marks):
    print(f"{name} scored {mark}")
```

```
# Week 2 - While Loops and Loop Control
# Basic while loop
count = 1
while count <= 5:
    print(f"Count is {count}")
    count += 1

# while loop with break
print("\nUsing break:")
num = 0
while True:
    print(num)
    num += 1
    if num >= 5:
        break

# while loop with continue
print("\nUsing continue - skip even numbers:")
n = 0
while n < 10:
    n += 1
    if n % 2 == 0:
        continue
    print(n)

# guess the number simulation (no input needed)
print("\nSimulated guessing game:")
secret = 7
guesses = [3, 9, 5, 7]
attempt = 0
for guess in guesses:
    attempt += 1
    if guess == secret:
        print(f"Attempt {attempt}: Correct! The number was {secret}")
        break
    elif guess < secret:
        print(f"Attempt {attempt}: {guess} is too low")
    else:
        print(f"Attempt {attempt}: {guess} is too high")
```

```
# Week 2 - Nested Loops
# Multiplication table using nested loops
print("Multiplication table (1-5):")
for i in range(1, 6):
    for j in range(1, 6):
        print(i * j, end="\t")
    print()

# Pattern printing - right triangle
print("\nRight triangle pattern:")
for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()

# Pattern printing - number pyramid
print("\nNumber pyramid:")
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()

# Nested loop over a list of lists (like a mini grid / table)
students_data = [
    ["Aastha", 85, 90],
    ["Riya", 78, 82],
    ["Ranbir", 92, 88]
]

print("\nStudent scores grid:")
for student in students_data:
    for value in student:
        print(value, end=" | ")
    print()
```

```
# Week 2 - Loops with List Comprehensions
numbers = list(range(1, 21))

# Traditional loop way
squares_loop = []
for n in numbers:
    squares_loop.append(n ** 2)

# List comprehension way (same result, shorter)
squares_comp = [n ** 2 for n in numbers]

print("Loop result:", squares_loop[:5])
print("Comprehension result:", squares_comp[:5])
print("Are they equal?", squares_loop == squares_comp)

# Filtering with comprehension vs loop
evens_loop = []
for n in numbers:
    if n % 2 == 0:
        evens_loop.append(n)

evens_comp = [n for n in numbers if n % 2 == 0]
print("\nEvens (loop):", evens_loop)
print("Evens (comprehension):", evens_comp)

# Nested comprehension - flatten a grid
grid = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat_loop = []
for row in grid:
    for val in row:
        flat_loop.append(val)

flat_comp = [val for row in grid for val in row]
print("\nFlattened (loop):", flat_loop)
print("Flattened (comprehension):", flat_comp)
```

```
# Week 2 - My Own Loop Practice (Solutions)

# 1. Print all odd numbers from 1 to 30 using a for loop
print("Odd numbers 1-30:")
for i in range(1, 31):
    if i % 2 != 0:
        print(i, end=" ")

# 2. Print your name 5 times using a while loop
print("\n\nMy name 5 times:")
count = 0
while count < 5:
    print("Aastha")
    count += 1

# 3. A nested loop pattern - inverted triangle
print("\nInverted triangle pattern:")
for i in range(5, 0, -1):
    for j in range(i):
        print("*", end="")
    print()

# 4. Using enumerate() over favourite things
print("\nMy favourite things:")
favourites = ["biryani", "coding", "music", "travelling", "reading"]
for index, item in enumerate(favourites, start=1):
    print(f"{index}. {item}")

# 5. List comprehension - cubes from 1 to 10
cubes = [n ** 3 for n in range(1, 11)]
print("\nCubes 1-10:", cubes)
```
